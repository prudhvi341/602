## Question 5. How does the distribution of `salesprice` vary by make for cars made between 2010-2015?
- Use the 10 most common makes (your cleaned version of this column).  
- Create a graph that answers this question.

````python
import matplotlib.pyplot as plt
df_filtered = df[(df['year'] >= 2010) & (df['year'] <= 2015)]
top_makes = df_filtered['make'].value_counts().head(10).index.tolist()
df_top_makes = df_filtered[df_filtered['make'].isin(top_makes)]


plt.figure(figsize=(12,8))
plt.boxplot([df_top_makes[df_top_makes['make']==make]['sellingprice'] for make in top_makes], labels=top_makes)
plt.title('Distribution of Sales Price by Make (2010-2015)')
plt.xlabel('Make')
plt.ylabel('Selling Price')
plt.show()
````
